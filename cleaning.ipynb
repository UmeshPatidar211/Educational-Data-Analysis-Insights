### Filling Null Value's

Filling Null value's in "ParentEdue" Column With it's Mode (Most Repeatitive Value)

```python
df['ParentEduc'].fillna(df['ParentEduc'].mode()[0], inplace=True)
```

```python
df['ParentEduc'].isnull().sum()
```

In "TestPrep" Replace "None" to "Incomplete"

```python
df['TestPrep'] = df['TestPrep'].replace('none','incomplete')
```

Fill All The Categorical Column's With It's With The Help Of Loop

```python
df.columns
```

```python
df.dtypes
```

```python
cat_columns = ['ParentEduc', 'LunchType', 'TestPrep', 'ParentMaritalStatus',
                'PracticeSport', 'IsFirstChild', 'TransportMeans', 'WklyStudyHours']
```

```python
cat_columns
```

```python
for i in df.columns:
    if i in cat_columns:
        df[i].fillna(df[i].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['NrSiblings'] = df['NrSiblings'].astype(int)
```
